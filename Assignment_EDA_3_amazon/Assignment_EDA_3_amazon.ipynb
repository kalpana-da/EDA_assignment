```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

import warnings 
warnings.filterwarnings('ignore')
# Load the dataset
df = pd.read_csv("amazon.csv")
df 
```

```python

df['discounted_price'] = df['discounted_price'].replace({'₹': '', ',': ''}, regex=True).astype(float)

df['discounted_price'].dtypes
```

```python
df['actual_price'] = df['actual_price'].replace({'₹': '', ',': ''}, regex=True).astype(float)
df['actual_price'].dtypes
```

```python
df['discount_percentage'] = df['discount_percentage'].replace({'%': ''}, regex=True).astype(float)
df['discount_percentage'] .dtypes
```


### Q1: What is the average selling price for each dealer, and how does it compare across different dealers?

**Insights:**
- The average selling prices for different dealers will show the price variation.
- Some dealers may offer premium products with higher prices, while others may focus on affordable options.


```python

df['discounted_price'] = df['discounted_price'].replace({'₹': '', ',': ''}, regex=True).astype(float)
df['actual_price'] = df['actual_price'].replace({'₹': '', ',': ''}, regex=True).astype(float)
category_price_avg = df.groupby('category')[['discounted_price', 'actual_price']].mean()
category_price_avg.head()
```

```python
# Plot the average prices for the top 10 categories
top_categories_sorted = category_price_avg.sort_values('actual_price', ascending=False).head(10)

# Plot the average discounted and actual prices for the top 10 categories
top_categories_sorted.plot(kind='bar', figsize=(12, 6), title="Top 10 Categories by Average Selling Price")
plt.xlabel('Category')
plt.ylabel('Price (₹)')
plt.xticks(rotation=45, ha='right')
plt.legend(title="Price Type", labels=['discounted_price', 'actual_price'])
plt.show()
```


### Q2: What is the average rating for each product category?

**Insights:**
>> The highest-rated categories are typically those that have a higher consumer satisfaction, while lower-rated categories may indicate areas that require improvement.
>>  Categories with the highest ratings include products like Office Supplies and Arts & Crafts, which generally have customer satisfaction scores above 4.0.
>>  Products with lower ratings (below 4) might need better quality control, customer service, or enhancements.


```python
plt.figure(figsize=(12, 6))
sns.barplot(x=top_categories_sorted.index, y=top_categories_sorted.values, palette='viridis')
plt.title("Top 10 Categories by Average Rating")
plt.xlabel('Category')
plt.ylabel('Average Rating')
plt.xticks(rotation=45, ha='right')
plt.show()

```


### Q3: What are the top rating_count products by category?

**Insights:**
- The products with the highest rating counts within each category will indicate higher customer engagement and popularity.


```python
# Identify the product with the highest rating count in each category
top_rated_products_by_category = df.groupby('category')['rating_count'].idxmax()

# Get the products corresponding to the highest rating counts
top_rated_products = df.loc[top_rated_products_by_category, ['category', 'product_name', 'rating_count']]

# Display the result
top_rated_products
```

```python

top_rated_products_by_category = df.groupby('category')['rating_count'].idxmax()
top_rated_products = df.loc[top_rated_products_by_category, ['product_name', 'rating_count']]

top_rated_products = top_rated_products.head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x='rating_count', y='product_name', data=top_rated_products, palette='viridis')
plt.title("Top 10 Products by Rating Count")
plt.xlabel('Rating Count')
plt.ylabel('Product Name')
plt.show()
```


### Q4: What is the distribution of discounted prices vs. actual prices?

**Insights:**
- A visualization of the distribution of both discounted and actual prices will show how products are priced across the market and how discounts affect prices.


```python

# Visualize the distribution of discounted and actual prices
plt.figure(figsize=(10, 6))
sns.histplot(df['discounted_price'], color='blue', label='Discounted Price', kde=True)
sns.histplot(df['actual_price'], color='red', label='Actual Price', kde=True)
plt.legend()
plt.title('Distribution of Discounted Prices vs. Actual Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```


### Q5: How does the average discount percentage vary across categories?

**Insights:**
- The discount percentage across categories will help us understand how discounting strategies differ by product category.


```python
# Remove non-numeric characters like '%' from 'discount_percentage' and convert to float
df['discount_percentage'] = df['discount_percentage'].replace({'%': '', ',': ''}, regex=True).astype(float)

# Now calculate the average discount percentage per category
category_discount_avg = df.groupby('category')['discount_percentage'].mean()

# Plot the top 10 categories with the highest average discount percentages
top_discount_categories = category_discount_avg.sort_values(ascending=False).head(10)

# Horizontal bar plot for the top 10 categories
plt.figure(figsize=(10, 6))
sns.barplot(x=top_discount_categories.values, y=top_discount_categories.index, palette='coolwarm')
plt.title("Top 10 Categories by Average Discount Percentage")
plt.xlabel('Average Discount Percentage (%)')
plt.ylabel('Category')
plt.show()
```


### Q6: What are the most popular product names?

**Insights:**
- By identifying the most frequent product names, we can understand which products are in high demand.


```python

# Find the most popular product names based on frequency
popular_product_names = df['product_name'].value_counts().head(10)
popular_product_names

# Plot the most popular product names
popular_product_names.plot(kind='barh', figsize=(10, 6), title="Most Popular Product Names")
plt.xlabel('Frequency')
plt.ylabel('Product Name')
plt.show()
```


### Q7: What are the most popular product keywords?

**Insights:**
- Extracting and counting the most frequent keywords from product names will help identify popular features and trends.


```python
# Extract and count keywords from product names
keywords = df['product_name'].str.extractall(r'(\w+)').unstack().stack().value_counts().head(10)

# Plot the most popular keywords
keywords.plot(kind='barh', figsize=(10, 6), title="Most Popular Product Keywords")
plt.xlabel('Frequency')
plt.ylabel('Keyword')
plt.show()

```


### Q8: What are the most popular product reviews?

**Insights:**
- Analyzing the most frequent review titles will help us understand the key themes and aspects that customers are focusing on in their feedback.


```python

# Find the most popular review titles based on frequency
popular_reviews = df['review_title'].value_counts().head(10)
popular_reviews

# Plot the most popular reviews
popular_reviews.plot(kind='barh', figsize=(12, 8), title="Most Popular Product Reviews")
plt.xlabel('Frequency')
plt.ylabel('Review Title')
plt.show()
```


### Q9: What is the correlation between discounted price and rating?

**Insights:**
- This analysis will show if there is any relationship between offering discounts and customer ratings.


```python
# Remove non-numeric characters from 'discounted_price' and 'rating' columns
df['discounted_price'] = pd.to_numeric(df['discounted_price'].replace({'₹': '', ',': ''}, regex=True), errors='coerce')
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# Now calculate the correlation between discounted price and rating
correlation = df[['discounted_price', 'rating']].corr()

# Plot the correlation
sns.heatmap(correlation, annot=True, cmap='coolwarm', center=0)
plt.title("Correlation Between Discounted Price and Rating")
plt.show()
```


### Q10: What are the Top 5 categories based on the highest ratings?

**Insights:**
- By identifying the top 5 categories with the highest ratings, we can highlight the best-performing areas in the product catalog.


```python

# Identify the top 5 categories with the highest average ratings
category_highest_rated = df.groupby('category')['rating'].mean().sort_values(ascending=False).head(5)
category_highest_rated

# Plot the top 5 categories by ratings
category_highest_rated.plot(kind='bar', figsize=(10, 6), title="Top 5 Categories by Highest Ratings")
plt.xlabel('Category')
plt.ylabel('Average Rating')
plt.xticks(rotation=90)
plt.show()
```
