# What is the average selling price of cars for each dealer, and how does it compare across different dealers?
# insights : U-Haul CO and Classic Chevy have the highest average prices (~$28,000+), likely offering newer or premium vehicles.
# Most dealers have similar average prices, suggesting a consistent pricing trend across the market.
# A few dealers, like Buddy Storbeck's Diesel Service, have lower average prices, possibly focusing on used or budget models.
# Overall, price differences are not drastic, indicating a balanced competitive market.

```python
#the average selling price of cars for each dealer
avg_price_per = df.groupby('Dealer_Name')['Price ($)'].mean().sort_values(ascending = False)
avg_price_per_dealer = round(avg_price_per, 0)
avg_price_per_dealer
```

```python
avg_price_per_dealer.plot(kind = 'barh', figsize = (10, 6), color='skyblue')
plt.ylabel(" Delear Name")
plt.xlabel(" Average Price")
plt.title("Average Selling Price per Dealer")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

## Q . 7) What is the distribution of car prices by region, and how does the number of cars sold vary by region?
# Insights : Regions with the highest car sales don't always have the highest average prices. This may suggest that in high-volume regions, more affordable cars are sold in bulk, while premium pricing appears in more selective regions.

```python
price_stats = df.groupby('Dealer_Region')['Price ($)'].agg(['mean', 'std']).reset_index()
price_stats.columns = ['Region', 'Avg Price', 'Price Std Dev']
price_stats
```

```python
car_count = df['Dealer_Region'].value_counts().reset_index()
car_count.columns = ['Region', 'Cars Sold']
car_count.columns
```

```python
region_summary = pd.merge(price_stats, car_count, on='Region')
region_summary.sort_values(by='Cars Sold', ascending=False, inplace=True)
region_summary
```

```python

fig, ax1 = plt.subplots(figsize=(10, 6))
ax1.bar(region_summary['Region'], region_summary['Cars Sold'], color='skyblue', label='Cars Sold')
ax1.set_ylabel('Cars Sold', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')
plt.xticks(rotation=45)
ax2 = ax1.twinx()
ax2.plot(region_summary['Region'], region_summary['Avg Price'], color='red', marker='o', label='Avg Price')
ax2.set_ylabel('Average Price ($)', color='red')
ax2.tick_params(axis='y', labelcolor='red')

plt.title('Cars Sold and Average Price by Region')
plt.tight_layout()
plt.show()
```

# Q. 12) Is there any seasonal trend in car sales based on the date of sale?
# Insights : You might see spikes in sales during certain months, like March, June, or December — often related to festive seasons, financial year-end, or holiday bonuses.
# Sales dips in months like February or August are also common in some regions due to weather, holidays, or low consumer spending periods.

```python

df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year
df['Month_Name'] = df['Date'].dt.strftime('%b')
monthly_sales = df.groupby('Month_Name').size().reindex(
    ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
     'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
).reset_index(name='Sales Count')

# Plot
plt.figure(figsize=(10, 5))
sns.barplot(data=monthly_sales, x='Month_Name', y='Sales Count', palette='viridis')
plt.title('Seasonal Trend: Car Sales by Month')
plt.xlabel('Month')
plt.ylabel('Number of Cars Sold')
plt.tight_layout()
plt.show()
```
